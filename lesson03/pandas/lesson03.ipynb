Index has been added by defult, since we have not specified the argument index_col while loading the file. Having had some knowledge about the file beforehand, we could have load the file using the following:

```python
df_alternative=pd.read_csv("data\\train.csv", index_col="PassengerId")
df_alternative
```

or

```python
df_alternative=pd.read_csv("data\\train.csv", index_col=0)
df_alternative
```

You have to know that you can do anything with that table. You can rename, drop, replace, slice, extract, apply functions, cut... What we will show you is just a fraction of what is possible.

To rename columns, you have to create a dictionary.

```python
newNames={"Pclass": "PassengerClass", "Name": "FullName"}
```

```python
df_alternative=df_alternative.rename(mapper=newNames, axis=1)
df_alternative
```

To drop columns, you have to create a list of them (unless you want to drop a single column).

```python
colstoDrop=["Cabin"]
```

```python
df_alternative=df_alternative.drop(labels=colstoDrop, axis=1)
df_alternative
```

## PART 1.2

```python
df=pd.read_csv("data\\train.csv", index_col=0)
df
```

### We can access single values of a DataFrame:

```python
df["Name"][1]
```

```python
df.Name[1]
```

#### We can also do it more efficiently (and maybe intuitively?):

```python
df.loc[1, "Name"]
```

```python
df.iloc[0,2]
```

#### Can you see the difference between loc and iloc?

#### What you will use more than accessing is slicing. Also here, there are a few ways to do that and you should be extremely careful with how you index stuff.

```python
df.iloc[:5,:]
```

```python
df.iloc[:5, :2]
```

```python
df.iloc[:5, :20]
```

```python
df.iloc[-5:,:]
```

```python
df.iloc[2:3,:]
```

```python
df.loc[2:3,:]
```

```python
df.loc[:, "Survived":"Pclass"]
```

```python
df.loc[3:6, "Survived":"Pclass"]
```

```python
df.iloc[2:6, 0:2]
```

```python
df.loc[3:6, ["Survived", "Pclass", "Age"]]
```

#### Just a final remark - creating own DataFrames is possible. We will not cover that, can happen though if you get data in another disjoint format like .txt files, .xml or .json.
