# Filtering DataFrames

#### We can create a boolean series from a DataFrame column...

```python
df["points"] > 90
```

#### ... and use it to extract (filter) the table. Here we are interested on all the rows, for which the number of points exceeded 90.

```python
df[df["points"] > 90]
```

#### We can also assign the filtering condition to a variable and filter by it...

```python
mask = df["points"] > 90

df[mask]
```

#### ... or by its negation using "~"

```python
df[~mask]
```

#### If we want to allow for more values, we can create a list and check if an element is in it...

```python
countries_lst=["Italy", "Germany"]

df["country"].isin(countries_lst)
```

#### ... and then filter by it

```python
df[df["country"].isin(["Italy", "Germany"])]
```

#### We can also combine the conditions using boolen operations...

```python
mask1 = df["points"] > 90
mask2 = df["country"].isin(["Italy", "Germany"])
```

#### AND

```python
df[mask1 & mask2]
```

#### or OR ;)

```python
df[mask1 | mask2]
```

#### or even filter a single column based on another. Here we are interested in the regions scoring more than 90 points.

```python
mask = df["points"] > 90

df["region_1"][mask]
```

#### We can also modify a column based on another. Here we want to make all Italian wines 5 units more expensive.

```python
mask = (df["country"]=="Italy")

df["price"][mask] += 5
df
```

# Transforming DataFrames

#### We can take advantage of pandas built-in function, NumPy functions working on a whole vector or overwritten operators.

```python
df["NewPrice"]=df["price"]+5
df
```

#### We can also write our own functions for the single element and apply it element-wise (more common). For that we can use the member function of the DataFrame type: .apply().

```python
def add_5(el):
    return el+5
```

```python
df["price"].apply(add_5)
```

#### Additional: lambda objects are a nice way to write short, one-lined functions, which we can then use directly as an argument.

```python
df["price"].apply(lambda x: x+5)
```

#### Remark: If possible, apply the built-in functions working on vectors. The "apply" function works as a loop and is usually less performative. Of course, there is only a limited number of vectorized operations - if yours is not available, feel free to use .apply(). It is still better than a "for" loop. 

#### We can also combine columns or define columns based on the others. Here we want to have a name of the winery which always includes a country first.

```python
df["country"]+": "+df["winery"]
```

#### Remember that just applying a function on a column does not change anything in your DataFrame. You have to assign the result to a column, so that your DataFrame gets updated.

# Groupby

#### We group our DataFrame according to the country...

```python
obj_grouped=df.groupby("country", sort=False)
```

#### ... and count the rows for each country.

```python
obj_grouped.count()
```

#### We can be also interested in seeing the first row of each of the groups...

```python
obj_grouped.first()
```

#### ... or calculate the average values for each of the countries (note for which columns the average was computed)

```python
obj_grouped.mean()
```

#### You can also use groupby with multiple keys and apply your own functions to groups.
