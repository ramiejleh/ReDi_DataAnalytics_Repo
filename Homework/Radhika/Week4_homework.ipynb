```python
import pandas as pd
import numpy as np
filename = 'data/recipes.csv'
df = pd.read_csv(filename, index_col = 0)
df
```

#### Print the names of Columns that have null values

```python
df.columns[df.isnull().any()].tolist()
```

#### Print the recipes with the highest number of aggregateLikes

```python

df_liked_recipes = df.groupby('title')['aggregateLikes'].sum().sort_values(ascending = False)
df_liked_recipes.idxmax()
```

#### Handle the missing values of “calcium percent of Daily need”. Sort the resulting DataFrame by id column.

```python
df_missing_values = df.loc[df['CalciumpercentOfDailyNeed'].isna()]

df_missing_values.sort_values(by = ['id'])
```

#### Delete the columns that have null values == 141

```python

col_to_drop = [col for col in df.columns if df[col].isnull().sum() == 141]
df.drop(col_to_drop, axis=1)
```

#### Print the title of the recipes that are vegetarian == TRUE and glutenFree == True

```python
df_filter_recipes = df[(df.vegetarian == True) & (df.glutenFree == True)]
df_filter_recipes.title
```

#### How many vegan recipes are there (Vegan = Vegetarian and dairy free)?

```python
df_filter_recipes = df[(df.vegetarian == True) & (df.dairyFree == True)]
df_filter_recipes
df_filter_recipes.title.value_counts().sum()
```

#### Compare the average amount of Vitamin B12 for the vegan and non-vegan recipes. How reliable are the results?

```python
vitaminB12_vegan_recipes = df[df.vegetarian == True]
vegan_vitaminB12 = vitaminB12_vegan_recipes.VitaminB12Amount.mean() 

vitaminB12_non_vegan_recipes = df[df.vegetarian == False]
non_vegan_vitaminB12 = vitaminB12_non_vegan_recipes.VitaminB12Amount.mean()

print('Avg amount of Vitamin B12 for vegan recipes: ' + str(vegan_vitaminB12))
print('Avg amount of Vitamin B12 for vegan recipes: ' + str(non_vegan_vitaminB12))
```

#### Compare the average health Score for both types of recipes (vegan vs non-vegan).

```python
healthScore_vegan = df[df.vegetarian == True].healthScore.mean()

healthScore_vegan
```

```python
healthScore_non_vegan = df[df.vegetarian == False].healthScore.mean()
healthScore_non_vegan
```
