#### Handle the missing values of “calcium percent of Daily need”. Sort the resulting DataFrame by id column.

```python
df_receipe['CalciumpercentOfDailyNeed'].isnull().sum()
```

```python
df_receipe['CalciumpercentOfDailyNeed'].fillna(float(df_receipe['CalciumpercentOfDailyNeed'].mean()), inplace=True)
```

```python
df_receipe['CalciumpercentOfDailyNeed'].isnull().sum()
```

```python
df_receipe.sort_values(by='id',inplace=True)
df_receipe.head()
```

#### Delete the columns that have null values == 141

```python
col = df_receipe[df_receipe.columns[df_receipe.isnull().sum()==141]]
col
```

```python
df_receipe1=df_receipe.drop(columns=col)
df_receipe1.isnull().sum()
```

#### Print the title of the recipes that are vegetarian == TRUE and glutenFree == True

```python
mask1 = df_receipe1['vegetarian']
mask2 = df_receipe1['glutenFree']

df_receipe1[mask1&mask2]
```

#### How many vegan recipes are there (Vegan = Vegetarian and dairy free)?

```python
mask3 = df_receipe1['dairyFree']

vegan_recipes = pd.DataFrame(df_receipe1[mask1&mask3])
vegan_recipes.shape
```

```python
vegan = df_receipe1.groupby(['dairyFree','vegetarian']).count()
vegan
```
