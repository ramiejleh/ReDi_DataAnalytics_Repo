```python
## Read the file into memory:

import csv
file = 'pension_fund_data-Phuong.csv'

with open(file) as raw_data:
    reader = csv.reader(raw_data, delimiter=';')
    header = next(reader)
       
    fund_data = []
    for row in reader:
        fund_data.append(row)
    header[1] = header[1].replace(' ','')
        
```

```python
## Remove any wrong or odd row entries:

fund_names = set(row[0] for row in fund_data)
print(fund_names)

for i, row in enumerate(fund_data):
    if row[0] == 'Nofund':
        #print(fund_data[i])
        del fund_data[i]
        
#company_names = set(row[1] for row in fund_data)
#sorted_company_names = sorted(company_names)
#for company_name in sorted_company_names:
    # print(company_name)
    
#invested_amounts = set(row[2] for row in fund_data)
#print(invested_amounts)

#for row in fund_data:
    #if row[0] == 'Nofund':
        #print (reader.line_num, row)
```

```python
## Remove 'company_name: ' in company name collumn:

def remove_text(string):
    return string.replace('company_name: ','')

for row in fund_data:
    row[1] = remove_text(row[1])
    
#print(fund_data)

#remove_text("company_name: hihi")
#def remove_text(string, remove_text):
    #return string.replace(remove_text,'')
```

```python
## Formatting the invested amounts (please don't run this cell)
import math

for row in fund_data:
    if row[0] == 'AkademikerPension':
        invested_round_up = math.ceil(float(row[2]))
        row[2] = invested_round_up
    elif row[0] == 'Industriens Pension':
        millions_formatted = int(float(row[2]) * 1000000)
        row[2] = millions_formatted
                
```

```python
## Write clean data to a new file:

with open('Pension fund - Clean data.csv','w', newline = '\n') as clean_data:
    writer = csv.writer(clean_data)
    writer.writerow(header)
    writer.writerows(fund_data)
```
