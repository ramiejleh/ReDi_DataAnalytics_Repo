Let's do some plotting :) 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
df = pd.read_csv("data/company_sales_data.csv")
df
```

### Exercise 1: Show 'total_profit' as a line plot with 'month_number' on x-axis

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(df.month_number, df.total_profit, color="red")
ax.set_xlabel("Month")
ax.set_ylabel("Total Profit")
```

```python
plt.plot(df.month_number, df.total_profit)
plt.ylabel("Total Profit")
plt.xlabel("Month")
```

### Exercise 2: Show 'total_profit' as a line plot with 'month_number' on x-axis, but...
- The line style must be dashed
- The line color must be red
- The line width must be 3
- A scatter plot has to be placed on top of the line (with color = 'black')
- A legend has to be placed in the lower right corner (remember that this requires a label)

```python
plt.plot(df.month_number, df.total_profit, linestyle="dashed", color="red", linewidth=3, marker="o", 
         markerfacecolor='black') #, label="Total Profit")
plt.scatter(df.month_number, df.total_profit, color="black")
plt.ylabel("Total Profit")
plt.xlabel("Month")
plt.legend(["Total Profit"], loc="lower right")
```

### Exercise 3: Show sales data for the following products in the same plot (separate lines for each product)
- facecream
- facewash
- toothpaste
- bathingsoap
- shampoo
- moisturizer

```python
plt.plot(df.month_number, df.facecream, color="red", label="facecream")
plt.plot(df.month_number, df.facewash, color="blue", label="facewash")
plt.plot(df.month_number, df.toothpaste, color="purple", label="toothpaste")
plt.plot(df.month_number, df.bathingsoap, color="black", label="bathingsoap")
plt.plot(df.month_number, df.shampoo, color="yellow", label="shampoo")
plt.plot(df.month_number, df.moisturizer, color="green", label="moisturizer")
plt.legend(loc="upper left")
```

```python
# df[["facecream", "facewash", "toothpaste", "bathingsoap", "shampoo", "moisturizer"]].plot()
df.plot(x="month_number", y=["facecream", "facewash", "toothpaste", "bathingsoap", "shampoo", "moisturizer"])
plt.legend(loc="upper left")
```

```python
for var in ["facecream", "facewash", "toothpaste", "bathingsoap", "shampoo", "moisturizer"]:
    plt.plot(df["month_number"], df[var], label=var)
    
plt.legend(loc="upper left")
```

### Exercise 4: Calculate the correlation between 'facewash' and 'moisturizer', and visualize the result in a scatterplot 

```python
x = "facewash"
y = "toothpaste"
print(f"correlation = {np.corrcoef(df[x], df[y])[0][1]}")

plt.scatter(x=df[x], y=df[y])
b, a = np.polyfit(df[x], df[y], deg=1)
plt.plot(df[x], df[x]*b+a, color="k")
```

### Exercise 5: Plot the correlations between all columns in the dataset using Seaborn's heatmap (sns.heatmap)

```python
corrmat= df.corr()
plt.figure(figsize=(9,5))  
sns.heatmap(corrmat,annot=True, cmap="BuPu", center=0)
```
