#### Handle the missing values of ‚Äúcalcium percent of Daily need‚Äù. Sort the resulting DataFrame by id column.

```python
value_to_fill = df.CalciumpercentOfDailyNeed.median()
new_df = df.copy()
new_df.CalciumpercentOfDailyNeed.fillna(value_to_fill, inplace=True)
new_df.sort_values(by=['id'], inplace=True)
new_df
```

#### Delete the columns that have null values == 141

```python
col_to_drop = [col for col in df.columns if df[col].isnull().sum() == 141]
df.drop(col_to_drop, axis=1)
```

#### How many vegan recipes are there (Vegan = Vegetarian and dairy free)?

```python
vegan = df[(df.vegetarian == True) & (df.dairyFree == True)]
vegan.title.count()
```

#### Compare the average amount of Vitamin B12 for the vegan and non-vegan recipes. How reliable are the results?

```python
non_vegan = df[(df.vegetarian == False) | (df.dairyFree == False)]
non_vegan.VitaminB12Amount.isnull().sum()
vegan.VitaminB12Amount.isnull().sum()
print(f"Average amount of Vitamin B12 for the vegan recipes: {vegan.VitaminB12Amount.mean()}")
print(f"Average amount of Vitamin B12 for the non-vegan recipes: {non_vegan.VitaminB12Amount.mean()}")
print(f"The result is not reliable because {vegan.VitaminB12Amount.isnull().sum()} vegan recipes are null values, in the total of {vegan.title.count()} vegan recipes")
```

#### Create a new column: "VeganWeek" where you decrease the price of vegan products by 10%, but only if they already cost more than 100.

```python
df['VeganWeek'] = np.where((df.vegetarian == True) & (df.dairyFree == True) & (df.pricePerServing > 100), df.pricePerServing*0.9, df.pricePerServing)
df
```
