#### Compare the average amount of Vitamin B12 for the vegan and non-vegan recipes. How reliable are the results?

```python
vitaminB12_vegan_recipes = df[df.vegetarian == True]
vegan_vitaminB12 = vitaminB12_vegan_recipes.VitaminB12Amount.mean() 

vitaminB12_non_vegan_recipes = df[df.vegetarian == False]
non_vegan_vitaminB12 = vitaminB12_non_vegan_recipes.VitaminB12Amount.mean()

print('Avg amount of Vitamin B12 for vegan recipes: ' + str(vegan_vitaminB12))
print('Avg amount of Vitamin B12 for vegan recipes: ' + str(non_vegan_vitaminB12))
```

```python
#check results' reliability 
#The results are not reliable as there are many missing observations, which could change the mean value. 
a= vitaminB12_non_vegan_recipes.VitaminB12Amount.isnull().count()
b= vitaminB12_vegan_recipes.VitaminB12Amount.isnull().count()
print('missing values of VitaminB12Amount in vegan and non-vegan recipes are ' + str(b) + ' and ' + str(a))
```
