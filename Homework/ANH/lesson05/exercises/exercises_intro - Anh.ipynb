```python
#Create a function to name seasons according to values in season column

def find_season_names(dataframe):
    if dataframe.season == 0.0:
        return 'spring'
    elif dataframe.season == 1.0:
        return 'summer'
    elif dataframe.season == 2.0:
        return 'autumn'
    else:
        return 'winter'
    
    
#Create season_names column and call the function 
df['season_names'] = df.apply(find_season_names, axis = 1)

#check season_names with different season column's values 
df.loc[df.season == 0.0]
```

#### What was the average number of bike rides in winter? How was the number in the summer? What can you say about these numbers and is a direct comparison the right thing to do? Why? Why not?

```python
#What was the average number of bike rides in winter? 

winter_bike_rides = df.loc[df.season == 3.0]['cnt'].mean()

#How was the number in the summer?
summer_bike_rides = df.loc[df.season == 1.0]['cnt'].mean()

print('Avg bikes rides in summer was {} and in winter it was {}'.format(int(summer_bike_rides), int(winter_bike_rides)))

#These numbers imply there were more cyclists in summer than those in winter in the surveyed period. 

#Count the number of row within winter periods to validate the mean comparison
df.loc[df.season == 3.0].count()
```

```python
#Count the number of row within summer periods to validate the mean comparison
df.loc[df.season == 1.0].count()

#There are 4330 and 4387 rows (only 57 rows different) for bikes rides in winter and summer periods, so the average values of bike rides in these two season are both acceptable. 
```
