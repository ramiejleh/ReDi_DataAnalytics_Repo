```python
data['Marital_Status'] = data['Marital_Status'].replace(['Married', 'Together'],'Relationship')
data['Marital_Status'] = data['Marital_Status'].replace(['Divorced', 'Widow', 'Alone', 'YOLO', 'Absurd'],'Single')
```

```python
plt.figure(figsize=(28, 6))
sns.countplot(x="Marital_Status", data=data)
```

```python
plt.figure(figsize=(28, 6))
sns.distplot(data["Income"], color='Seagreen')
```

```python
data['NumTotalPurchases'] = data['NumWebPurchases'] + data['NumCatalogPurchases'] + data['NumStorePurchases'] + data['NumDealsPurchases']
```

```python
plt.figure(figsize=(28, 6))
sns.distplot(data["NumTotalPurchases"],color = 'Seagreen')
```

```python
data['Expenditure'] = data['MntWines'] + data['MntFruits'] + data['MntMeatProducts'] + data['MntFishProducts'] + data['MntSweetProducts'] + data['MntGoldProds']
```

```python
plt.figure(figsize=(28, 6))
sns.distplot(data["Expenditure"],color = 'Seagreen')
```

```python
data['TotalAcceptedCmp'] = data['AcceptedCmp1'] + data['AcceptedCmp2'] + data['AcceptedCmp3'] + data['AcceptedCmp4'] + data['AcceptedCmp5']
plt.figure(figsize=(28, 6))
sns.distplot(data["TotalAcceptedCmp"],color = 'Seagreen')
```

```python
col_to_del = ["AcceptedCmp1" , "AcceptedCmp2", "AcceptedCmp3" , "AcceptedCmp4","AcceptedCmp5","NumWebVisitsMonth", "NumWebPurchases","NumCatalogPurchases","NumStorePurchases","NumDealsPurchases" , "Kidhome", "Teenhome","MntWines", "MntFruits", "MntMeatProducts", "MntFishProducts", "MntSweetProducts", "MntGoldProds"]
data=data.drop(columns=col_to_del,axis=1)
print(data.columns)
data.head()
```

```python
from datetime import date
data['Dt_Customer'] = pd.to_datetime(data.Dt_Customer)
data['First_day'] = '01-01-2015'
data['First_day'] = pd.to_datetime(data.First_day)
data['Day_engaged'] = (data['First_day'] - data['Dt_Customer']).dt.days
```

```python
data['Age'] = (pd.Timestamp('now').year) - (pd.to_datetime(data['Dt_Customer']).dt.year)
print("Unique categories present in the Age:\n",data['Age'].value_counts())
data['Age'].value_counts().plot(kind='bar',color = 'seagreen' ,linewidth = 3)
plt.title("Frequency Of Each Category in the Age Variable \n")
```
