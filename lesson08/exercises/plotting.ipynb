In Python, we have three majorly adopted plotting packages:
- Matplotlib
- Seaborn
- Plotly

Today, we want to show how each package can be used for plotting.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
df = pd.read_csv("data/company_sales_data.csv")
df.head(2)
```

### Exercise 1: Show 'total_profit' as a line plot with 'month_number' on x-axis

```python
plt.plot(df["month_number"], df["total_profit"])
plt.ylabel("Total Profit")
plt.xlabel("Month")
```

### Exercise 2: Show 'total_profit' as a line plot with 'month_number' on x-axis, but...
- The line style must be dashed
- The line color must be red
- The line width must be 3
- A scatter plot has to be placed on top of the line (with color = 'black')
- A legend has to be placed in the lower right corner (remember that this requires a label)

```python
plt.plot(df["month_number"], df["total_profit"], linestyle='--', color='red', linewidth=3, label="Total Profit")
plt.scatter(x=df["month_number"], y=df["total_profit"], color='black')
plt.ylabel("Total Profit")
plt.xlabel("Month")
plt.legend(loc="lower right")
```

### Exercise 3: Show sales data for the following products in the same plot (separate lines for each product)
- facecream
- facewash
- toothpaste
- bathingsoap
- shampoo
- moisturizer

```python
plt.figure(figsize=(10,6))

for col in ["facecream", "facewash", "toothpaste", "bathingsoap", "shampoo", "moisturizer"]:
    plt.plot(df["month_number"], df[col], label=col)
plt.xlabel("Month")
plt.legend(loc='upper left')
```

### Exercise 4: Calculate the correlation between 'facewash' and 'moisturizer', and visualize the result in a scatterplot 

```python
x = 'facewash'
y = 'moisturizer'

print('correlation = ', np.corrcoef(df[x], df[y])[0][1])

# fit a linear model
coef = np.polyfit(df[x], df[y], 1)
best_fit = coef[1] + coef[0] * df[x]

plt.plot(df[x], best_fit, color='k')
plt.scatter(x=df[x], y=df[y])
plt.xlabel(x)
plt.ylabel(y)
plt.show()
```

### Exercise 5: Plot the correlations between all columns in the dataset using Seaborn's heatmap (sns.heatmap)

```python
sns.heatmap(df.corr(), annot=True, cmap="BuPu", center=0)
```
